````python
POINTS = {
    "q01": 1, "q02": 1, "q03": 1, "q04": 1, "q05": 1,
    "q06": 1, "q07": 1, "q08": 1, "q09": 1, "q10": 1,
    "q11": 1, "q12": 1, "q13": 1, "q14": 1, "q15": 1,
    "q16": 1, "q17": 1, "q18": 1, "q19": 1, "q20": 1,
    "qb1": 3, "qb2": 3, "qp1": 8, "qp2": 8, "qp3": 8,
}
TOTAL_POINTS = sum(POINTS.values())

scores = [grader.check(q) for q in [
    "q01","q02","q03","q04","q05","q06","q07","q08","q09","q10",
    "q11","q12","q13","q14","q15","q16","q17","q18","q19","q20",
    "qb1","qb2","qp1","qp2","qp3"]]
passed = sum(1 for s in scores if s.grade == 1)
score = sum(POINTS[s.name] for s in scores if s.grade == 1)
summary = ", ".join(f"{s.name}:{'OK' if s.grade == 1 else 'NG'}" for s in scores)
print(summary)
print(f"合格: {passed} / 25 問")
print(f"得点: {score} / {TOTAL_POINTS} 点")

import hashlib
name = input("氏名を入力してください: ")
token = hashlib.sha256(f"{name}-{summary}".encode()).hexdigest()[:12]
print("=" * 40)
print(f"氏名: {name}")
print(f"合格数: {passed}")
print(f"得点: {score}")
print(f"合否サマリー: {summary}")
print(f"提出コード: {token}")
print("氏名・合格数・得点・合否サマリー・提出コードの5つをフォームに貼り付けてください")
````
