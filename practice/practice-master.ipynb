# 練習おわり

3問とも ✅ になれば、本番の受け方はバッチリです。
最後に下のセルを実行して、提出コードの表示まで体験しておきましょう。

```python
POINTS = {"p01": 1, "pb1": 2, "pp1": 3}
TOTAL_POINTS = sum(POINTS.values())

scores = [grader.check(q) for q in ["p01", "pb1", "pp1"]]
passed = sum(1 for s in scores if s.grade == 1)
score = sum(POINTS[s.name] for s in scores if s.grade == 1)
summary = ", ".join(f"{s.name}:{'OK' if s.grade == 1 else 'NG'}" for s in scores)
print(summary)
print(f"正解: {passed} / 3 問")
print(f"得点: {score} / {TOTAL_POINTS} 点")

import hashlib
name = input("氏名を入力してください: ")
token = hashlib.sha256(f"{name}-{summary}".encode()).hexdigest()[:12]
print("=" * 40)
print(f"氏名: {name}")
print(f"正解数: {passed}")
print(f"得点: {score}")
print(f"解答結果: {summary}")
print(f"提出コード: {token}")
print("（練習なので提出は不要です）")
```
